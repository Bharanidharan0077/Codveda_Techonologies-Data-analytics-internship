# EDA Analysis - Customer Churn Dataset

# Import Libraries

```python
import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

#For Better visulization
sns.set(style="whitegrid")
```

# Load Dataset

```python
df = pd.read_csv("C:/Users/dhine/Desktop/Codveda Techonologies/Cleaned_churn-bigml-80.csv")
```

# Basic informatiion

```python
#Shape of dataset
print("Shape:", df.shape)

#Data types
df.info()

#Statistical summary
df.describe()
```

# Checking null values

```python
df.isnull().sum()
```

```python
# Target variables Analysis(churn)
```

```python
#Count plot
sns.countplot(x='Churn', data=df)
plt.title("Churn distribution")
plt.show()

#percentage
df['Churn'].value_counts(normalize=True) * 100
```

# Categorical features Analysis

```python
# International Plan
sns.countplot(x="International plan", hue="Churn", data=df)
plt.title("Churn vs International Plan")
plt.show()
```

```python
sns.countplot(x='Voice mail plan', hue='Churn', data=df)
plt.title("Churn vs Voice Mail Plan")
plt.show()
```

# Numerical Feature Analysis

```python
#Distribution plot
df.hist(figsize=(15,10))
plt.show()
```

```python
#Box plot(Outlier Detection)
plt.figure(figsize=(10,5))
sns.boxplot(data=df[['Total day minutes','Total eve minutes','Total night minutes']])
plt.title("Usage Distribution")
plt.show()
```

# Customer Service calls vs churn

```python
sns.countplot(x='Customer service calls', hue='Churn', data=df)
plt.title("Customer service calls vs churn")
plt.show()
```

# Correlation Heatmap

```python
plt.figure(figsize=(10,8))
corr=df.corr(numeric_only=True)

sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Correlation Matric")
plt.show()
```

# Key Relationship Analysis

```python
# Day minutes vs churn
sns.boxplot(x='Churn', y='Total day minutes', data=df)
plt.title("Day minutes vs churn")
plt.show()
```

```python
#Charges vs churn
sns.boxplot(x="Churn", y="Total day charge", data=df)
plt.title("Day charge vs Churn")
plt.show()
```

```python
#pairplot
sns.pairplot(df[['Total day minutes','Total eve minutes','Total night minutes','Churn']], hue='Churn')
plt.show()
```

## Key Insights

- Customers with high customer service calls are more likely to churn
- International plan users have higher churn rate
- High day usage and charges increase churn probability
- Voice mail plan users churn less

## Conclusion

- Customer behavior and service interaction are major churn drivers
- Business should focus on high-risk customers to improve retention
